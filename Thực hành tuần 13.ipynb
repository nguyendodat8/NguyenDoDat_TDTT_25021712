```python
#W11A1
file = open('songuyen.txt','w')
so_nguyen = input()
file.write(so_nguyen)
file.close
def reverse(so_nguyen):
    so_nguyen = int(so_nguyen)
    cso = list()
    while so_nguyen > 0:
        chu_so = so_nguyen % 10
        cso.append(chu_so)
        so_nguyen //= 10
    cso.sort(reverse=True)
    S = 0
    for i in range(len(cso)):
        S += cso[i] * 10 ** (len(cso) - 1 - i)
    return S
print(reverse(so_nguyen))
```

```python
#W11A12
import os

def maximumProduct(path: str) -> int:
    # Đọc dữ liệu từ file
    with open(path, 'r') as f:
        numbers = list(map(int, f.read().split()))
    
    # Sắp xếp dãy số
    numbers.sort()
    
    # Lấy ba số lớn nhất
    max1 = numbers[-1]
    max2 = numbers[-2]
    max3 = numbers[-3]
    
    # Lấy hai số nhỏ nhất
    min1 = numbers[0]
    min2 = numbers[1]
    
    # Trường hợp 1: tích của 3 số lớn nhất
    product1 = max1 * max2 * max3
    
    # Trường hợp 2: tích của 2 số nhỏ nhất và số lớn nhất
    product2 = min1 * min2 * max1
    
    # Kết quả cuối cùng
    return max(product1, product2)

```

```python
#W11A16
import math
import os

def tinh_chu_vi():
    file_path = input("Nhập đường dẫn đến tệp tin: ")

    if not os.path.exists(file_path):
        print("Lỗi: Tệp tin không tồn tại!")
        return

    try:
        with open(file_path, 'r') as f:
            for line in f:
                parts = line.split()
                if not parts:
                    continue
                
                ten_hinh = parts[0].lower()
                
                try:
                    if ten_hinh == "vuong":
   
                        canh = float(parts[1])
                        chu_vi = 4 * canh
                        
                    elif ten_hinh == "tron":
                       
                        ban_kinh = float(parts[1])
                        chu_vi = 2 * math.pi * ban_kinh
                        
                    elif ten_hinh == "chunhat":
                       
                        dai = float(parts[1])
                        rong = float(parts[2])
                        chu_vi = 2 * (dai + rong)
                    else:
                        continue
                    
                    print(f"{chu_vi:.2f}")
                    
                except (ValueError, IndexError):
                    print("Lỗi dữ liệu dòng:", line.strip())
                    
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

if __name__ == "__main__":
    tinh_chu_vi()
```

```python
#W11A17
import os
def productExceptSelf(path):
    with open (path,'r') as f:
        n = int(f.readline().strip())
        arr = list(map(int, f.readline().split()))
        result = []
        S = 0
        for i in range (n):
            if i == 0:
                S *= arr[(i+1):n-1]
                result.append(S)
            elif i == n-1:
                S *= arr[0:(n-2)]
                result.append(S)
            else:
                S *= arr[:i] + arr[i:]
                result.append(S)
    return S
path = input()
print(productExceptSelf(path)) 
```

```python
#W11A18
import math

def estimatedTime(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    

    src_parts = lines[0].strip().split()
    x1 = float(src_parts[2])
    y1 = float(src_parts[4])
    

    des_parts = lines[1].strip().split()
    x2 = float(des_parts[2])
    y2 = float(des_parts[4])
    

    velocity = float(lines[2].strip().split()[1])
    

    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    

    time = distance / velocity
    
    return round(time, 2)
path = input()
print(estimatedTime(path))
```

```python
#W11A20
import math
def imagrSmoother(path:str):
    try:
        # 1. Đọc dữ liệu từ file
        with open(path, 'r') as f:
            lines = f.readlines()
        
        if not lines:
            return
            
        # Dòng đầu tiên là kích thước M
        M = int(lines[0].strip())
        
        # Đọc ma trận ảnh
        img = []
        for i in range(1, M + 1):
            row = list(map(int, lines[i].split()))
            img.append(row)
            
        # 2. Khởi tạo ma trận kết quả
        # Chúng ta cần tạo ma trận mới để không làm thay đổi giá trị gốc khi đang tính toán
        rows = len(img)
        cols = len(img[0]) if rows > 0 else 0
        result = [[0] * cols for _ in range(rows)]
        
        # 3. Duyệt từng điểm ảnh
        for r in range(rows):
            for c in range(cols):
                total_sum = 0
                count = 0
                
                # Duyệt các ô lân cận trong phạm vi 3x3
                for i in range(r - 1, r + 2):
                    for j in range(c - 1, c + 2):
                        # Kiểm tra xem tọa độ có nằm trong ma trận không
                        if 0 <= i < rows and 0 <= j < cols:
                            total_sum += img[i][j]
                            count += 1
                
                # Tính trung bình và làm tròn xuống (floor division)
                result[r][c] = total_sum // count
        
        # 4. In kết quả ra màn hình
        for row in result:
            print(*(row))
            
    except FileNotFoundError:
        print("Không tìm thấy tệp tin.")
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

# Ví dụ cách gọi hàm:
# imageSmoother('input.txt')
```
