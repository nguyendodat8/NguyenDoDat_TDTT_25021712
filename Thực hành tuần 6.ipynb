```python
#W6A5
def count_numbers(numbers):
    # Initialize the dictionary with zero counts
    counts = {
        'positives': 0,
        'negatives': 0,
        'zeros': 0
    }
    
    for num in numbers:
        if num > 0:
            counts['positives'] += 1
        elif num < 0:
            counts['negatives'] += 1
        else:
            counts['zeros'] += 1
            
    return counts
```

```python
#W6A9
dict1 = eval(input("Nhập dictionary thứ nhất: "))
dict2 = eval(input("Nhập dictionary thứ hai: "))
merged_dict = {}
for key, value in dict1.items():
    merged_dict[key] = merged_dict.get(key, 0) + value
for key, value in dict2.items():
    merged_dict[key] = merged_dict.get(key, 0) + value
sorted_dict = dict(sorted(merged_dict.items()))
print("Dictionary sau khi gộp:", sorted_dict)

```

```python
#W6A16
n, m = map(int, input("Nhập n và m: ").split())
matrix = []

# Nhập mảng
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

# In mảng định dạng 4 vị trí
for row in matrix:
    for x in row:
        print(f"{x:4d}", end="") # :4d giúp căn phải 4 khoảng trống
    print() # Xuống dòng sau mỗi hàng
```

```python
#W6A17
n = int(input("Nhập kích thước n: "))
matrix = [list(map(int, input().split())) for _ in range(n)]

main_diagonal = []
side_diagonal = []

for i in range(n):
    main_diagonal.append(matrix[i][i])
    side_diagonal.append(matrix[i][n-1-i])

print("Đường chéo chính:", main_diagonal)
print("Đường chéo phụ:", side_diagonal)
```

```python
#W6A18
line = list(map(int, input().split()))
n, m, k = line[0], line[1], line[2]

matrix = [list(map(int, input().split())) for _ in range(n)]

# Tính tổng trên cột k (giả sử k nhập vào bắt đầu từ chỉ số 0)
tong_cot = 0
for i in range(n):
    tong_cot += matrix[i][k]

# Tính tổng trên hàng k
tong_hang = sum(matrix[k])

print(f"Tổng cột {k}: {tong_cot}")
print(f"Tổng hàng {k}: {tong_hang}")
```

```python
#W6A19
def vi_tri_max(lst):
    max_val = lst[0]
    vi_tri = 1

    for i in range(1, len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
            vi_tri = i + 1   # +1 vì vị trí bắt đầu từ 1

    return vi_tri
lst = list(map(int,input().split()))
print(vi_tri_max(lst))
```

```python
#W6A20
def vi_tri_dau_tien(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i + 1   # vị trí bắt đầu từ 1
    return -1
lst = list(map,int(input().split()))
k = int(input())
print(vi_tri_dau_tien(lst,k))
```
