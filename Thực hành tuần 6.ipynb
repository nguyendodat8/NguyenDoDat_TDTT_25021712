```python
#W6A5
def count_numbers(numbers):
    counts = {
        'positives': 0,
        'negatives': 0,
        'zeros': 0
    }
    
    for num in numbers:
        if num > 0:
            counts['positives'] += 1
        elif num < 0:
            counts['negatives'] += 1
        else:
            counts['zeros'] += 1
            
    return counts
numbers = list(map(int,input().split()))
print(count_numbers(numbers))
```

```python
#W6A13
data = input().split()

# Tạo dictionary ban đầu
d = {}
for i in range(0, len(data), 2):
    key = data[i]
    value = data[i + 1]
    d[key] = value

# Đảo ngược key và value
new_dict = {}
for key, value in d.items():
    new_dict[value] = key

# In kết quả
print(new_dict)
```

```python
#W6A19
def vi_tri_max(lst):
    max_val = lst[0]
    vi_tri = 1

    for i in range(1, len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
            vi_tri = i + 1   # +1 vì vị trí bắt đầu từ 1

    return vi_tri
lst = list(map(int,input().split()))
print(vi_tri_max(lst))
```

```python
#W6A20
def vi_tri_dau_tien(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i + 1   # vị trí bắt đầu từ 1
    return -1
lst = list(map,int(input().split()))
k = int(input())
print(vi_tri_dau_tien(lst,k))
```

```python
#W6A21
line = input()
words = line.split()
pos_dict = {}
for index, word in enumerate(words):
    if word not in pos_dict:
        pos_dict[word] = (index,)
    else:
        pos_dict[word] = pos_dict[word] + (index,)
print(pos_dict)
```
