```python
def swap(a,b):
    return b,a
a,b = map(int,input().split())
print(swap(a,b))
```

```python
import math
def is_prime(x):
    if x<2:
        return False
    if x == 2:
        return True
    else:
        for i in range(2,int(math.sqrt(x)+1)):
            if x%i == 0:
                return False
            return True
x = int(input())
print(is_prime(x))
    
```

```python
def la_so_hoan_hao(n):
    S = 0
    for i in range (1,n):
        if n%i == 0:
            S += i
    if S == n:
        return True
    return False
n = int(input())
print(la_so_hoan_hao(n))
```

```python
def k_in_list(k,lst):
    for i in range (0,len(lst)):
        if lst[i]==k:
            return i+1
    return -1
lst = list(map(int,input().split()))
k = int(input())
print(k_in_list(k,lst))
```

```python
def Tinh_giai_thua(n):
    if n<=0:
        return None
    else:
        S = 1
        for i in range (1,n+1):
            S *= i
        return S
n = int(input())
print(Tinh_giai_thua(n))
```

```python
def May_tinh_bo_tui(a,b,x):
    if x == "+":
        return round(a+b,2)
    if x == "-":
        return round(a-b,2)
    if x == "*":
        return round(a*b,2)
    if x == "/":
        return round(a/b,2)
    return None
a = float(input())
b = float(input())
x = input()
print(May_tinh_bo_tui(a,b,x))
```

```python
def Tong_chu_so(n):
    if n<=0:
        return None
    else:
        tong = 0
        while n > 0:
            chu_so = n % 10
            tong += chu_so
            n //=10
        return tong
n = int(input())
print(Tong_chu_so(n))
```

```python
def khoang_cach_hamming(a,b):
    m = bin(a^b)
    cnt = 0
    for i in m:
        i = 1
        cnt += 1
    return cnt
a,b = map(int,input()).split()
print(khoang_cach_hamming(a,b))
```

```python
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False
    
    map_a_to_b = {}
    map_b_to_a = {}
    
    for char_a, char_b in zip(a, b):
        # Kiểm tra ánh xạ từ a -> b
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
        
        # Kiểm tra ánh xạ từ b -> a
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a
    
    return True
```
