```python
def Max_of_Two(a,b):
    return max(a,b)
a= int(input())
b= int(input())
print(Max_of_Two(a,b))
```

```python
def swap(a,b):
    return b,a
a,b = map(int,input().split())
print(swap(a,b))
```

```python
def is_prime(x):
    if x<2:
        return False
    if x == 2:
        return True
    else:
        for i in range(2,x+1):
            if x%i == 0:
                return False
            return True
x = int(input())
print(is_prime(x))
    
```

```python
def la_so_hoan_hao(n):
    S = 0
    for i in range (1,n):
        if n%i == 0:
            S += i
    if S == n:
        return True
    return False
n = int(input())
print(la_so_hoan_hao(n))
```

```python
def k_in_list(k):
    for i in range (1,len(lst)):
        if list[i]==k:
            return i
    return -1
lst = list(map(int,input().split()))
k = int(input())
print(k_in_list(k))
```

```python
def k_in_ds(k):
    if k in ds:
        return k
    return -1
k = int(input())
ds = map(int,input().strip().split())
print(k_in_ds(k))

```

```python
def Tinh_giai_thua(n):
    if n<=0:
        return None
    else:
        S = 1
        for i in range (1,n+1):
            S *= i
        return S
n = int(input())
print(Tinh_giai_thua(n))
```

```python
def k_in_list(k, lst):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

lst = list(map(int, input().split()))
k = int(input())
print(k_in_list(k, lst))

```

```python
def May_tinh_bo_tui(a,b,x):
    if x == "+":
        return round(a+b,2)
    if x == "-":
        return round(a-b,2)
    if x == "*":
        return round(a*b,2)
    if x == "/":
        return round(a/b,2)
    return None
a = float(input())
b = float(input())
x = input()
print(May_tinh_bo_tui(a,b,x))
```

```python
def Tong_chu_so(n):
    if n<0:
        return None
    else:
        tong = 0
        while n > 0:
            chu_so = n % 10
            tong += chu_so
            n //=10
        return tong
n = int(input())
print(Tong_chu_so(n))
```
