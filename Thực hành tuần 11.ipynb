```python
#W11A3
def so_hanh_phuc(a):
    da_gap = set()
    while a != 1 and a not in da_gap:
        da_gap.add(a)
        S = 0
        while a > 0:
            chu_so = a % 10
            S += chu_so ** 2
            a //= 10
        a = S
    return a == 1
try:
    a = int(input())
    if so_hanh_phuc(a):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("Dữ liệu nhập vào cần là 1 số")
```

```python
#W11A10
def kiem_tra_hop_le(ds):
    hop_le = True
    for i in ds:
        for j in range(len(ds)):
            for k in range (i,len(ds)):
                if ds[j] == ds[k]:
                    hop_le = False
                    return "Mảng không hợp lệ"
            hop_le = True
            S = 0
            for i in ds:
                S += i
            return S
try:
    ds = list(map(int,input().split()))
    print(kiem_tra_hop_le(ds))
except:
    print("Mảng không hợp lệ")

    
```
