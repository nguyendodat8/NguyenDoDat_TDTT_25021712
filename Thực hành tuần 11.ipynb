```python
#W11A6
import math
def square_x(x):
    return round(math.sqrt(x),2)
try:
    x = int(input())
    print(square_x(x))
except ValueError:
    print('So am')
```

```python
#W11A7
def tuoi_hop_le(tuoi):
    if tuoi >= 0:
       nam_sinh = 2025 - tuoi
       return nam_sinh
try:
    tuoi = int(input())
    if tuoi >= 0:
        print(tuoi_hop_le(tuoi))
    else:
        print("Tuổi không hợp lệ")
except (ValueError,TypeError):
    print("Tuổi không hợp lệ")
```

```python
#W11A10
def kiem_tra_hop_le(ds):
    hop_le = True
    for i in ds:
        for j,k in range(len(ds)):
            if ds[j] == ds[k]:
                hop_le = False
            hop_le = True
            S = 0
            for i in ds:
                S += i
    return S
try:
    ds = list(map(int,input().split()))
    print(kiem_tra_hop_le(ds))
except:
    print("Mảng không hợp lệ")
    
```
