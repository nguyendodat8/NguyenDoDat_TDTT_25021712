```python
#W8A1
class Cylinder:
    def __init__(self,r,h):
        self.r = r
        self.h = h
    def Dien_tich(self,r,h,pi=3.14):
        Dien_tich = 2*pi*self.r*self.h + 2*pi*self.r*self.r
        DT = round(Dien_tich,2)
        return DT
    def The_tich(self,r,h,pi=3.14):
        The_tich = pi*self.r*self.r*self.h
        TT = round(The_tich,2)
        return TT
    def __repr__ (self,DT,TT):
        return f" Diện tích: {DT} , Thể tích: {TT}"
    
r = float(input())
h = float(input())
Cylinder1 = Cylinder(r,h)
print(Cylinder1.Dien_tich(r,h))
print(Cylinder1.The_tich(r,h))    
```

```python
#W8A2
class Date:
    def __init__(self,day,month,year):
        self.day = day
        self.month = month
        self.year = year
    def kiem_tra_hop_le(self,day,month,year):
        if self.month < 1 or self.month > 12:
            return False
        else:
            if self.month in [1,3,5,7,8,10,12]:
                if self.day < 1 or self.day > 31:
                    return False
            elif self.month in [4,6,9,11]:
                if self.day < 1 or self.day > 30:
                    return False
            elif self.month == 2:
                if (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0):
                    if self.day < 1 or self.day > 29:
                        return False
                else:
                    if self.day < 1 or self.day > 28:
                        return False
            return True
    def tinh_ngay_ke_tiep(self,day,month,year):
        if self.kiem_tra_hop_le(day,month,year) == False:
            return "Ngày không hợp lệ"
        else:
            ngay = self.day
            thang = self.month
            nam = self.year
            ngay += 1
            if thang in [1,3,5,7,8,10]:
                if ngay > 31:
                    ngay = 1
                    thang += 1
            elif thang in [4,6,9,11]:
                if ngay > 30:
                    ngay = 1
                    thang += 1
            elif thang == 2:
                if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
                    if ngay > 29:
                        ngay = 1
                        thang += 1
                else:
                    if ngay > 28:
                        ngay = 1
                        thang += 1
            if thang == 12:
                ngay = 1
                thang = 1
                nam += 1
            return f"{ngay}/{thang}/{nam}"
day = int(input())
month = int(input())
year = int(input())
Date1 = Date(day,month,year)
print(Date1.kiem_tra_hop_le(day,month,year))
print(Date1.tinh_ngay_ke_tiep(day,month,year))
```

```python
#W8A7
class Fraction:
    def __init__ (self,tu,mau):
        self.tu = tu
        self.mau = mau
    def add(self,tu,mau):
        tu_moi = self.tu*mau + self.mau*tu
        mau_moi = self.mau*mau
        for i in range(2,min(tu_moi,mau_moi)+1):
            if tu_moi % i == 0 and mau_moi % i == 0:
                tu_moi = tu_moi // i
                mau_moi = mau_moi // i
        return f"Tử mới: {tu_moi} , Mẫu mới: {mau_moi}"
    def subtract(self,tu,mau):
        tu_moi = self.tu*mau - self.mau*tu
        mau_moi = self.mau*mau
        for i in range(2,min(tu_moi,mau_moi)+1):
            if tu_moi % i == 0 and mau_moi % i == 0:
                tu_moi = tu_moi // i
                mau_moi = mau_moi // i
        return f"Tử mới: {tu_moi} , Mẫu mới: {mau_moi}"
    def multiply(self,tu,mau):
        tu_moi = self.tu*tu
        mau_moi = self.mau*mau
        for i in range(2,min(tu_moi,mau_moi)+1):
            if tu_moi % i == 0 and mau_moi % i == 0:
                tu_moi = tu_moi // i
                mau_moi = mau_moi // i
        return f"Tử mới: {tu_moi} , Mẫu mới: {mau_moi}"
    def divide(self,tu,mau):
        tu_moi = self.tu*mau
        mau_moi = self.mau*tu
        for i in range(2,min(tu_moi,mau_moi)+1):
            if tu_moi % i == 0 and mau_moi % i == 0:
                tu_moi = tu_moi // i
                mau_moi = mau_moi // i
        return f"Tử mới: {tu_moi} , Mẫu mới: {mau_moi}"
tu = int(input())
mau = int(input())
Fraction1 = Fraction(tu,mau)
print(Fraction1.add(tu,mau))
print(Fraction1.subtract(tu,mau))
print(Fraction1.multiply(tu,mau))
print(Fraction1.divide(tu,mau))     
```

```python
#W8A8:
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    def deposit(self,amount):
        self.balance += amount
        return self.balance
    def withdraw(self,amount):
        if amount > self.balance:
            return None
        else:
            self.balance -= amount
            return self.balance
    def __str__(self):
        return f"{self.owner},{self.balance}"
owner = input()
balance = float(input())
BankAccount1 = BankAccount(owner,balance)
n = int(input())
for i in range(1,n+1):
    action = input()
    amount = float(input())
    if action == 'deposit':
        print(f"DEPOSIT {amount}")
        print(BankAccount1.deposit(amount))
    elif action == 'withdraw':
        print(f"WITHDRAW {amount}")
        print(BankAccount1.withdraw(amount))
print(BankAccount1)    
        
```

```python
#W8A10
class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
        self.book = []
class Library(Book):
    def __init__(self,title,author,year):
        super().__init__(title,author,year)
    def add_book(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
        self.book.append({self.title:[self.author,self.year]})
    def count_by_author(self,author):
        self.author = author
        count = 0
        for i in self.book:
            for j,k in i.items():
                if k[0] == self.author:
                    count += 1
        return count
    def count_by_year(self,year):
        self.year = year
        count = 0
        for i in self.book:
            for j,k in i.items():
                if k[1] == self.year:
                    count += 1
        return count
n = int(input())
Library1 = Library("", "", 0)
for i in range(1,n+1):
    lenh = input()
    if lenh == 'add':
        title = input()
        author = input()
        year = int(input())
        Library1.add_book(title,author,year)
        print(f"ADD {title} {author} {year}")
    if lenh ==  'count by author':
        author = input()
        print(f"COUNT{author}")
        print(Library1.count_by_author(author))
    if lenh == 'count by year':
        year = int(input())
        print(f"COUNT {year}")
        print(Library1.count_by_year(year))
```

```python
#W8A11:
class Point():
    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z
        self.point = []
    def add_point(self,x,y,z):
        for i in range (1,5):
            self.x = x
            self.y = y
            self.z = z
            self.point.append((self.x,self.y,self,z))
    def tao_2_vector(self,p1,p2):
        self.p1 = self.point[p1]
        self.p2 = self.point[p2]
        vector = (self.p2[0]-self.p1[0],self.p2[1]-self.p1[1],self.p2[2]-self.p1[2])
        return vector
    def lap_phuong_trinh_mat_phang(self,p1,p2,p3):
        self.p1 = self.point[p1]
        self.p2 = self.point[p2]
        self.p3 = self.point[p3]
        v1 = (self.p2[0]-self.p1[0],self.p2[1]-self.p1[1],self.p2[2]-self.p1[2])
        v2 = (self.p3[0]-self.p1[0],self.p3[1]-self.p1[1],self.p3[2]-self.p1[2])
        a = v1[1]*v2[2] - v1[2]*v2[1]
        b = v1[2]*v2[0] - v1[0]*v2[2]
        c = v1[0]*v2[1] - v1[1]*v2[0]
        d = -(a*self.p1[0] + b*self.p1[1] + c*self.p1[2])
        return f"{a}x + {b}y + {c}z + {d} = 0"
    def goc_giua_hai_mat_phang(self,p1,p2,p3,p4):
        self.p1 = self.point[p1]
        self.p2 = self.point[p2]
        self.p3 = self.point[p3]
        self.p4 = self.point[p4]
        v1 = (self.p2[0]-self.p1[0],self.p2[1]-self.p1[1],self.p2[2]-self.p1[2])
        v2 = (self.p4[0]-self.p3[0],self.p4[1]-self.p3[1],self.p4[2]-self.p3[2])
        dot_product = v1[0]*v2[0] + v1[1]*v2[1] + v1[2]*v2[2]
        magnitude_v1 = (v1[0]**2 + v1[1]**2 + v1[2]**2)**0.5
        magnitude_v2 = (v2[0]**2 + v2[1]**2 + v2[2]**2)**0.5
        import math
        cos_theta = dot_product / (magnitude_v1 * magnitude_v2)
        angle = math.acos(cos_theta)
        angle_degrees = math.degrees(angle)
        return round(angle_degrees,2)
x = float(input())
y = float(input())
z = float(input())
Point1 = Point(x,y,z)
print(Point1.goc_giua_hai_mat_phang(0,1,2,3))
```
