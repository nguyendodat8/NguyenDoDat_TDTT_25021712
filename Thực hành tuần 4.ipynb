```python
import sys
import math

def largest_prime_factor(n: int) -> int:
    # Lấy ước tố 2 trước
    max_pf = -1
    while n % 2 == 0:
        max_pf = 2
        n //= 2

    # Kiểm tra các ước lẻ từ 3,5,7,... đến sqrt(n)
    i = 3
    # sử dụng i*i <= n để cập nhật theo n đã bị chia nhỏ
    while i * i <= n:
        while n % i == 0:
            max_pf = i
            n //= i
        i += 2

    # Nếu còn n > 1 thì n là một số nguyên tố lớn nhất
    if n > 1:
        max_pf = n

    return max_pf

def main():
    data = sys.stdin.read().strip().split()
    if not data:
        return
    # Giả sử mỗi dòng là một test (theo yêu cầu trước đây),
    # nhưng nếu chỉ một số, chương trình vẫn chạy.
    out_lines = []
    for token in data:
        n = int(token)
        if n < 2:
            out_lines.append("Input phải >= 2")
        else:
            out_lines.append(str(largest_prime_factor(n)))
    print("\n".join(out_lines))

if __name__ == "__main__":
    main()
```

```python
def is_palindrome(number):
    """
    Kiểm tra xem một số có phải là số đối xứng (palindrome) hay không.
    """
    str_num = str(number)
    return str_num == str_num[::-1]

def reverse_number(number):
    """
    Đảo ngược các chữ số của một số nguyên.
    """
    return int(str(number)[::-1])

def find_palindrome_steps(n):
    """
    Lặp lại phép toán n = n + reverse(n) cho đến khi n trở thành số đối xứng.
    Trả về số bước và giá trị palindrome thu được.
    """
    steps = 0
    while not is_palindrome(n):
        reversed_n = reverse_number(n)
        n += reversed_n
        steps += 1
    return steps, n

def main():
    """
    Hàm chính để nhập số nguyên dương và thực hiện các phép toán.
    """
    try:
        # Nhập số nguyên dương n
        n = int(input("Nhập số nguyên dương n: "))
        if n <= 0:
            print("Vui lòng nhập một số nguyên dương lớn hơn 0.")
            return
        
        # Tìm số bước và giá trị palindrome
        steps, palindrome = find_palindrome_steps(n)
        
        # In kết quả
        print(f"Số bước thực hiện: {steps}")
        print(f"Giá trị palindrome thu được: {palindrome}")
    
    except ValueError:
        print("Vui lòng nhập một số nguyên hợp lệ.")

if __name__ == "__main__":
    main()
```

```python
import math

def is_unique_digits(num):
    digits = str(num)
    return len(set(digits)) == len(digits)

def print_unique_square_numbers(n):
    result = []
    for i in range(1, int(math.sqrt(n)) + 1):
        square = i * i
        if is_unique_digits(square):
            result.append(str(square))
    if result:
        print(" ".join(result))
    else:
        print("no number")

# Nhập vào số nguyên dương n
try:
    n = int(input("Nhập số nguyên dương n: "))
    if n > 0:
        print_unique_square_numbers(n)
    else:
        print("Vui lòng nhập số nguyên dương.")
except ValueError:
    print("Đầu vào không hợp lệ.")

```

```python
def collatz_length(x):
    count = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        count += 1
    return count

def find_max_collatz(n):
    max_length = 0
    result = 1
    for i in range(1, n + 1):
        length = collatz_length(i)
        if length > max_length:
            max_length = length
            result = i
    print(f"{result} {max_length}")

# Nhập vào số nguyên dương n
try:
    n = int(input("Nhập số nguyên dương n: "))
    if n > 0:
        find_max_collatz(n)
    else:
        print("Vui lòng nhập số nguyên dương.")
except ValueError:
    print("Đầu vào không hợp lệ.")

```

```python
def count_even_divisors(n):
    count = 0
    for i in range(2, n + 1, 2):  # chỉ duyệt các số chẵn
        if n % i == 0:
            count += 1
    return count

# Nhập vào số nguyên dương n
try:
    n = int(input("Nhập số nguyên dương n (< 10^6): "))
    if 0 < n < 10**6:
        result = count_even_divisors(n)
        print(f"Số lượng ước số chẵn của {n} là: {result}")
    else:
        print("Vui lòng nhập số nguyên dương nhỏ hơn 10^6.")
except ValueError:
    print("Đầu vào không hợp lệ.")

```

```python
def tinh_tien_tiet_kiem(x, n):
    lai_suat = 0.007  # 0.7% mỗi tháng
    tong_tien = x * ((1 + lai_suat) ** n)
    return int(tong_tien)  # Bỏ phần lẻ thập phân

# Nhập dữ liệu từ người dùng
try:
    x = float(input("Nhập số tiền gửi ban đầu (X): "))
    n = int(input("Nhập số tháng gửi (N): "))
    if x > 0 and n > 0:
        so_tien = tinh_tien_tiet_kiem(x, n)
        print(f"Số tiền sau {n} tháng là: {so_tien} đồng")
    else:
        print("Vui lòng nhập số tiền và số tháng lớn hơn 0.")
except ValueError:
    print("Đầu vào không hợp lệ.")

```

```python
def tong_uoc_so(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong

def kiem_tra_than_thiet(a, b):
    return tong_uoc_so(a) == b and tong_uoc_so(b) == a

# Nhập hai số nguyên dương từ bàn phím
try:
    a = int(input("Nhập số thứ nhất: "))
    b = int(input("Nhập số thứ hai: "))
    if a > 0 and b > 0:
        if kiem_tra_than_thiet(a, b):
            print("true")
        else:
            print("false")
    else:
        print("Vui lòng nhập hai số nguyên dương.")
except ValueError:
    print("Đầu vào không hợp lệ.")
```

```python
def ucln(m, n):
    while n != 0:
        m, n = n, m % n
    return m

# Nhập hai số nguyên dương từ bàn phím
try:
    m = int(input("Nhập số nguyên dương m: "))
    n = int(input("Nhập số nguyên dương n: "))
    if m > 0 and n > 0:
        print(f"Ước chung lớn nhất của {m} và {n} là: {ucln(m, n)}")
    else:
        print("Vui lòng nhập hai số nguyên dương.")
except ValueError:
    print("Đầu vào không hợp lệ.")

```

```python
import string

def is_strong_password(password: str) -> bool:
    # Kiểm tra độ dài
    if len(password) <= 6:
        return False
    
    # Các điều kiện
    has_upper = any(ch.isupper() for ch in password)       # Có chữ in hoa
    has_lower = any(ch.islower() for ch in password)       # Có chữ thường
    has_digit = any(ch.isdigit() for ch in password)       # Có số
    has_special = any(ch in string.punctuation for ch in password)  # Có ký tự đặc biệt
    
    return has_upper and has_lower and has_digit and has_special





```
