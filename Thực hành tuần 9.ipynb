```python
arr1 = list(map(int,input().split()))
arr2 = list(map(int,input().split()))
sorted_array = []
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    for i in range(0,len(arr)):
        mid = len(arr)//2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])
        return merge(left,right)
def merge(left,right):
    merged_array = []
    i = j = 0
    while i<len(left) and j<len(right):
        if left[i] < right[j]:
            merged_array.append(left[i])
            i += 1
        merged_array.append(right[j])
        j += 1
    merged_array.extend(left[i:])
    merged_array.extend(right[j:])
    return merged_array
merged_array1 = merge_sort(arr1)
merged_array2 = merge_sort(arr2)
sorted_array.extend(merged_array1)
sorted_array.extend(merged_array2)
print(sorted_array)
```

```python
#W9A2
def tim_so_lon_thu_k(matrix,k):
    n,m = len(matrix),len(matrix[0])
    left,right = matrix([0][0]),matrix([n-1][m-1])
    def count_number(x):
        count = 0
        i,j = n-1,0
        while i>= 0 and j<m:
            if matrix[i][j] <= x:
                count += i+1
                j += 1
            else:
                i -= 1
        return count
    while left < right:
        mid = (left + right)//2
        if count_number(mid) < k:
            left = mid + 1
        else:
            right = mid
    return left
arr = list(map(int,input().split()))
matrix = []
index = 0
n = arr[0]
m = arr[1]
for i in range(n):
    row = []
    for j in range(m):
        row.append(arr[index + 2])
        index += 1
    matrix.append(row)
k = arr[-1]
print(tim_so_lon_thu_k(matrix,k))

```

```python
#W9A4
def xoa_chu_cai_thuong(a):
    for char in a:
        if char.islower():
            a = a.replace(char,'')
    return a
def chuyen_chu_thuong_sang_hoa(a):
    count = 0
    for char in a:
        if char.islower():
            count += 1
    for i in range (0,count+1):
        a = a.replace(a[i],a[i].upper())
        return a
Q = int(input())
for i in range(1,Q+1):
    a = input()
    b = input()
    if not b.isupper:
        break
    a = xoa_chu_cai_thuong(a)
    if a == b:
        print("YES")
    else:
        a = chuyen_chu_thuong_sang_hoa(a)
        if a == b:
            print("YES")
    print("NO")
```

```python
#W9A6:
def hoan_vi_tiep_theo(arr):
    n = len(arr)
    i = n - 2
    while i >= 0 and arr[i] >= arr[i + 1]:
        i -= 1
    if i == -1:
        return [-1]
    j = n - 1
    while arr[j] <= arr[i]:
        j -= 1
    arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1:] = reversed(arr[i + 1:])
    return arr
n = int(input())
arr = list(map(int,input().split()))
print(hoan_vi_tiep_theo(arr))
```

```python
#W9A7
def canPlaceFlowers(flowerbed,k):
    count = 0
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i-1] == 0)
            right_empty = (i == n-1) or (flowerbed[i+1] == 0)
            if left_empty and right_empty:
                flowerbed[i] = 1
                count += 1
    return count >= k
flowerbed = list(map(int,input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed,k))
```

```python
#W9A8
def is_path_crossing(moves):
    x,y = 0,0
    visited = set()
    visited.add((x,y))
    for move in moves:
        if move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        elif move == 'U':
            y += 1
        elif move == 'D':
            y -= 1
        if (x,y) in visited:
            return True
        visited.add((x,y))
    return False
move = list(map(str,input().split()))
print(is_path_crossing(move))
```

```python
#W9A9
def Median(nums):
    for i in range(len(nums)):
        nums[i] = nums[0]
        for j in range(i+1,len(nums)):
            if nums[j] < nums[i]:
                nums[i] = nums[j]
    n = len(nums)  
```

```python
#W9A10
def findRadius(houses,heaters):
    heaters.sort()
    radius = 0
    for house in houses:
        left,right = 0,len(heaters)-1
        while left <= right:
            mid = left + (right - left)//2
            if heaters[mid] < house:
                left = mid + 1
            else:
                right = mid -1
        if right >= 0:
            dist1 = abs(house - heaters[right])
        else:
            dist1 = float('inf')
        if left < len(heaters):
            dist2 = abs(heaters[left] - house)
        else:
            dist2 = float('inf')
        radius = max(radius,min(dist1,dist2))
    return radius
houses = list(map(int,input().split()))
heaters = list(map(int,input().split()))
print(findRadius(houses,heaters))
```
