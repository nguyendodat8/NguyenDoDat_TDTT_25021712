```python
#W7A1
def binary_search(arr,left,right,target):
    if left > right:
        return -1
    mid = (left+right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search(arr,left,mid-1,target)
    else:
        return binary_search(arr,mid+1,right,target)
arr = list(map(int,input().split()))
left = 0
right = len(arr) - 1
target = int(input())
print(binary_search(arr,left,right,target))        
```

```python
#W7A3
#Viết hàm thực hiện thuật toán bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range (0,n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1] , arr[j]
                sorted_arr = bubble_sort(arr)
    return sorted_arr
arr=list(map(int,input().split()))
#Viết hàm thực hiện sắp xếp lựa chọn:
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1,n):
            min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
arr=list(map(int,input().split()))
sorted_arr = selection_sort(arr)
print(sorted_arr)
```

```python
#W7A4
def phan_tu_nhieu_nhat(arr):
    count = 0
    max_count = 0
    result = arr[0]
    for number in arr:
        count = 0
        for x in arr:
            if x == number:
                count += 1
              
        if count > max_count:
            max_count = count
            result = number
    return result,max_count
arr = list(map(int,input().split()))
ket_qua, so_lan = phan_tu_nhieu_nhat(arr)
print(f"{ket_qua} xuat hien nhieu nhat,som nhat,{so_lan} lan")
```

```python
#W7A5
def dem_cap(arr, X):
    dem = 0  # biến đếm số cặp

    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == X:
                dem += 1
    return dem
arr = list(map(int,input().split()))
X = int(input())
print(dem_cap(arr,X))
```

```python
#W7A10
from bisect import bisect_right, insort

def max_subarray_modulo(nums, m):
    prefix = 0
    max_mod = 0
    prefix_list = []   # lưu các prefix đã gặp (luôn được sắp xếp)

    for num in nums:
        prefix = (prefix + num) % m

        # Cập nhật kết quả với prefix hiện tại
        max_mod = max(max_mod, prefix)

        # Tìm prefix nhỏ nhất > prefix hiện tại
        idx = bisect_right(prefix_list, prefix)

        if idx < len(prefix_list):
            candidate = (prefix - prefix_list[idx] + m) % m
            max_mod = max(max_mod, candidate)

        # Chèn prefix vào danh sách (giữ thứ tự)
        insort(prefix_list, prefix)

    return max_mod
```

```python
#W7A13
def day_hai_hoa(arr):
    n = len(arr)
    left = 0
    max_len = 0

    for right in range(n):
        # kiểm tra cửa sổ hiện tại
        while max(arr[left:right+1]) - min(arr[left:right+1]) > 1:
            left += 1

        max_len = max(max_len, right - left + 1)

    return max_len
arr = list(map(int,input().split()))
print(day_hai_hoa(arr))
```
